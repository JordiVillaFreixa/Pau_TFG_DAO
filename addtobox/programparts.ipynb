Ara demanem que ens crei un matriu de rotació 3d segons els angles de euller

```python
import numpy as np

def euler_rotation_matrix(phi, theta, psi):
    """
    Calcula la matriz de rotación en 3D según los ángulos de Euler (phi, theta, psi).
    
    Args:
        phi (float): Ángulo de Euler phi en radianes.
        theta (float): Ángulo de Euler theta en radianes.
        psi (float): Ángulo de Euler psi en radianes.
        
    Returns:
        numpy.ndarray: Matriz de rotación en 3D.
    """
    R_x = np.array([[1, 0, 0],
                    [0, np.cos(phi), -np.sin(phi)],
                    [0, np.sin(phi), np.cos(phi)]])
    
    R_y = np.array([[np.cos(theta), 0, np.sin(theta)],
                    [0, 1, 0],
                    [-np.sin(theta), 0, np.cos(theta)]])
    
    R_z = np.array([[np.cos(psi), -np.sin(psi), 0],
                    [np.sin(psi), np.cos(psi), 0],
                    [0, 0, 1]])
    
    # Combinar las tres rotaciones
    R = np.dot(R_z, np.dot(R_y, R_x))
    
    return R

# Ejemplo de uso
phi = np.pi/4  # Ángulo de Euler phi (en radianes)
theta = np.pi/3  # Ángulo de Euler theta (en radianes)
psi = np.pi/6  # Ángulo de Euler psi (en radianes)

# Calcular la matriz de rotación
R = euler_rotation_matrix(phi, theta, psi)

print("Matriz de rotación:")
print(R)
```

Demanem que ens fusioni els pdb en un únic, i que les separi uns 3 angstroms

```python
import os

def merge_pdbs(input_directory, output_file, separation_distance=3.0):
    with open(output_file, 'w') as out_pdb:
        atom_counter = 1
        for file_name in os.listdir(input_directory):
            if file_name.endswith('.pdb'):
                pdb_file = os.path.join(input_directory, file_name)
                with open(pdb_file, 'r') as in_pdb:
                    for line in in_pdb:
                        if line.startswith('ATOM'):
                            out_pdb.write(f"{line[:30]}{atom_counter:4}{line[34:]}")
                            atom_counter += 1
                # Añadir un espacio entre las moléculas
                out_pdb.write(f"TER\n")
                atom_counter = 1
                # Añadir separación entre las moléculas
                out_pdb.write(f"TER    {separation_distance:8.3f}   {separation_distance:8.3f}   {separation_distance:8.3f}\n")

# Directorio donde se encuentran los archivos PDB a fusionar
input_directory = "output_random"
# Nombre del archivo PDB de salida
output_file = "merged.pdb"

# Fusionar los archivos PDB en uno solo con separación de 3 Å entre moléculas
merge_pdbs(input_directory, output_file, separation_distance=3.0)

print(f"Archivos PDB fusionados en '{output_file}' con éxito.")
```
