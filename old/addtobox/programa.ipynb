```python
import numpy as np

def euler_rotation_matrix(phi, theta, psi):
    """
    Calcula la matriz de rotación en 3D según los ángulos de Euler (phi, theta, psi).
    
    Args:
        phi (float): Ángulo de Euler phi en radianes.
        theta (float): Ángulo de Euler theta en radianes.
        psi (float): Ángulo de Euler psi en radianes.
        
    Returns:
        numpy.ndarray: Matriz de rotación en 3D.
    """
    R_x = np.array([[1, 0, 0],
                    [0, np.cos(phi), -np.sin(phi)],
                    [0, np.sin(phi), np.cos(phi)]])
    
    R_y = np.array([[np.cos(theta), 0, np.sin(theta)],
                    [0, 1, 0],
                    [-np.sin(theta), 0, np.cos(theta)]])
    
    R_z = np.array([[np.cos(psi), -np.sin(psi), 0],
                    [np.sin(psi), np.cos(psi), 0],
                    [0, 0, 1]])
    
    # Combinar las tres rotaciones
    R = np.dot(R_z, np.dot(R_y, R_x))
    
    return R

# Ejemplo de uso
phi = np.pi/4  # Ángulo de Euler phi (en radianes)
theta = np.pi/3  # Ángulo de Euler theta (en radianes)
psi = np.pi/6  # Ángulo de Euler psi (en radianes)

# Calcular la matriz de rotación
R = euler_rotation_matrix(phi, theta, psi)

print("Matriz de rotación:")
print(R)
```

```python
import numpy as np

def read_pdb(file_path):
    coordinates = []
    with open(file_path, 'r') as pdb_file:
        for line in pdb_file:
            if line.startswith('ATOM'):
                x = float(line[30:38])
                y = float(line[38:46])
                z = float(line[46:54])
                coordinates.append([x, y, z])
    return np.array(coordinates)

def translate(coordinates, distance=4):
    """
    Translada las coordenadas para asegurar que la nueva molécula esté al menos a la distancia especificada
    del anterior.
    
    Args:
        coordinates (numpy.ndarray): Coordenadas cartesianas de la molécula.
        distance (float): Distancia mínima deseada entre las dos moléculas (por defecto 4 Å).
        
    Returns:
        numpy.ndarray: Coordenadas cartesianas trasladadas.
    """
    # Encontrar el centroide de las coordenadas existentes
    centroid = np.mean(coordinates, axis=0)
    
    # Calcular el vector de traslación multiplicando el vector unitario por la distancia deseada
    translation_vector = distance * np.random.rand(3)
    
    # Aplicar la traslación sumando el vector de traslación al centroide
    translated_coordinates = coordinates + translation_vector
    
    return translated_coordinates

def write_pdb(file_path, coordinates, chain_id):
    with open(file_path, 'w') as pdb_file:
        with open('tre-copy.pdb', 'r') as original_pdb:
            for line, coord in zip(original_pdb, coordinates):
                x_str = f"{coord[0]:8.3f}"
                y_str = f"{coord[1]:8.3f}"
                z_str = f"{coord[2]:8.3f}"
                new_line = f"{line[:21]}{chain_id}{line[22:30]}{x_str}{y_str}{z_str}{line[54:]}"
                pdb_file.write(new_line)

file_path = 'tre-copy.pdb'
coordinates = read_pdb(file_path)

print("Matriz de coordenadas cartesianas:")
print(coordinates)

# Definir ángulos de Euler
phi = np.pi / 4  # Ángulo de Euler phi (en radianes)
theta = np.pi / 3  # Ángulo de Euler theta (en radianes)
psi = np.pi / 6  # Ángulo de Euler psi (en radianes)

# Calcular la matriz de rotación
R = euler_rotation_matrix(phi, theta, psi)

# Aplicar la rotación a las coordenadas
rotated_coordinates = np.dot(R, coordinates.T).T

# Trasladar las coordenadas
translated_coordinates = translate(rotated_coordinates)

print("\nCoordenadas trasladadas:")
print(translated_coordinates)

# Escribir las coordenadas rotadas y trasladadas en un nuevo archivo PDB
write_pdb('rotated_translated.pdb', translated_coordinates, 'B')
```
